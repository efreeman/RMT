```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.stats.power as power

sns.set_style("whitegrid")
```

```python
p = power.TTestPower()

effect = [0.1, 0.3, 0.5, 0.7, 0.9]
sample = range(1, 100)
sig = 0.05

df = pd.DataFrame(columns=['Effect', 'Sample', 'Power'])

for e in effect:
    for n in sample:
        df = df.append({'Effect Size': e, 'Sample': n, 'Power': p.power(e, n, sig)}, ignore_index=True)

df['Effect Size'] = df['Effect Size'].astype('category')
```

```python
plot, axis = plt.subplots(figsize=(8, 5))
sns.lineplot(x=df['Sample'], y=df['Power'], hue=df['Effect Size'])
axis.set_xlabel("Sample Size")
axis.set_ylabel("Statistical Power")
```
